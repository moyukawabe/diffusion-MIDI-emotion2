def predict_evaluate(cuda,types,task,file_path)   
* types：default="midi_like", type=str, choices=["midi_like", "remi", "wav"]  
* task：default="ar_va", type=str, choices=["ar_va", "arousal", "valence"]  
* file_path：default=".", type=str  
* cuda：default='cuda:0,1', type=str  

```python
from inference_many import predict_evaluate # 'EMOPIA_cls/inference.py'
import numpy as np

def EMOPIA_cls_evaluate(file_path_Q1,file_path_Q2,file_path_Q3,file_path_Q4):
    cuda = 'cuda:2,3'
    types = "midi_like"
    task_V = "valence"
    task_A = "arousal"
    High_V_music_Q1 , Valence_label_Q1, Valence_values_Q1 = predict_evaluate(cuda,types,task_V,file_path_Q1) 
    High_A_music_Q1 , Arousal_label_Q1, Arousal_values_Q1 = predict_evaluate(cuda,types,task_A,file_path_Q1) 
    High_V_music_Q2 , Valence_label_Q2, Valence_values_Q2 = predict_evaluate(cuda,types,task_V,file_path_Q2) 
    High_A_music_Q2 , Arousal_label_Q2, Arousal_values_Q2 = predict_evaluate(cuda,types,task_A,file_path_Q2) 
    High_V_music_Q3 , Valence_label_Q3, Valence_values_Q3 = predict_evaluate(cuda,types,task_V,file_path_Q3) 
    High_A_music_Q3 , Arousal_label_Q3, Arousal_values_Q3 = predict_evaluate(cuda,types,task_A,file_path_Q3) 
    High_V_music_Q4 , Valence_label_Q4, Valence_values_Q4 = predict_evaluate(cuda,types,task_V,file_path_Q4) 
    High_A_music_Q4 , Arousal_label_Q4, Arousal_values_Q4 = predict_evaluate(cuda,types,task_A,file_path_Q4) 
    
    print("music estimate [V,A] ----------------") 
    print('Q1 \n valence \n ['+str(Valence_label_Q1)+', ] \n arousal \n ['+str(Arousal_label_Q1)+', ]')
    print('Q2 \n valence \n ['+str(Valence_label_Q2)+', ] \n arousal \n ['+str(Arousal_label_Q2)+', ]')
    print('Q3 \n valence \n ['+str(Valence_label_Q3)+', ] \n arousal \n ['+str(Arousal_label_Q3)+', ]')
    print('Q4 \n valence \n ['+str(Valence_label_Q4)+', ] \n arousal \n ['+str(Arousal_label_Q4)+', ]')
    
    
    print("music estimate value [V,A] ----------------") 
    print('Q1 \n valence \n ['+str(Valence_values_Q1)+', ] \n arousal \n ['+str(Arousal_values_Q1)+', ]')
    print('Q2 \n valence \n ['+str(Valence_values_Q2)+', ] \n arousal \n ['+str(Arousal_values_Q2)+', ]')
    print('Q3 \n valence \n ['+str(Valence_values_Q3)+', ] \n arousal \n ['+str(Arousal_values_Q3)+', ]')
    print('Q4 \n valence \n ['+str(Valence_values_Q4)+', ] \n arousal \n ['+str(Arousal_values_Q4)+', ]')
    
    print("average music estimate value [V,A] ----------------") 
    print('Q1 \n valence \n ['+str(np.mean(Valence_values_Q1, axis=0))+', ] \n arousal \n ['+str(np.mean(Arousal_values_Q1, axis=0))+', ]')
    print('Q2 \n valence \n ['+str(np.mean(Valence_values_Q2, axis=0))+', ] \n arousal \n ['+str(np.mean(Arousal_values_Q2, axis=0))+', ]')
    print('Q3 \n valence \n ['+str(np.mean(Valence_values_Q3, axis=0))+', ] \n arousal \n ['+str(np.mean(Arousal_values_Q3, axis=0))+', ]')
    print('Q4 \n valence \n ['+str(np.mean(Valence_values_Q4, axis=0))+', ] \n arousal \n ['+str(np.mean(Arousal_values_Q4, axis=0))+', ]')
    
    print("High music count ----------------") 
    print('Q1 ['+str(High_V_music_Q1)+', ] & ['+str(High_A_music_Q1)+', ]')
    print('Q2 ['+str(High_V_music_Q2)+', ] & ['+str(High_A_music_Q2)+', ]')
    print('Q3 ['+str(High_V_music_Q3)+', ] & ['+str(High_A_music_Q3)+', ]')
    print('Q4 ['+str(High_V_music_Q4)+', ] & ['+str(High_A_music_Q4)+', ]')
    
    print('----------------------------------------------')
    def print_VA(Q1_4,Valence,Arousal):
        print(str(Q1_4)+' = ['+str(Valence)+' , ' +str(Arousal)+'] &')
     
    # 感情の変化の向き
    
    Q2_Q1_V = np.mean(Valence_values_Q2 - Valence_values_Q1, axis=0)
    Q3_Q2_V = np.mean(Valence_values_Q3 - Valence_values_Q2, axis=0)
    Q4_Q3_V = np.mean(Valence_values_Q4 - Valence_values_Q3, axis=0)
    Q1_Q4_V = np.mean(Valence_values_Q1 - Valence_values_Q4, axis=0)
    Q2_Q1_A = np.mean(Arousal_values_Q2 - Arousal_values_Q1, axis=0)
    Q3_Q2_A = np.mean(Arousal_values_Q3 - Arousal_values_Q2, axis=0)
    Q4_Q3_A = np.mean(Arousal_values_Q4 - Arousal_values_Q3, axis=0)
    Q1_Q4_A = np.mean(Arousal_values_Q1 - Arousal_values_Q4, axis=0)
    #
    Q3_Q1_V = np.mean(Valence_values_Q3 - Valence_values_Q1, axis=0)
    Q4_Q2_V = np.mean(Valence_values_Q4 - Valence_values_Q2, axis=0)
    Q3_Q1_A = np.mean(Arousal_values_Q3 - Arousal_values_Q1, axis=0)
    Q4_Q2_A = np.mean(Arousal_values_Q4 - Arousal_values_Q2, axis=0)

     
    print('allmean_after_estimatevalue_difference ----------')   
    print_VA("Q1→Q2 [-,0] ",Q2_Q1_V,Q2_Q1_A)
    print_VA("Q2→Q3 [0,-] ",Q3_Q2_V, Q3_Q2_A)
    print_VA("Q3→Q4 [+,0] ",Q4_Q3_V, Q4_Q3_A)
    print_VA("Q4→Q1 [0,+] ",Q1_Q4_V, Q1_Q4_A)
    print_VA("Q1→Q3 [-,-] ",Q3_Q1_V,Q3_Q1_A)
    print_VA("Q2→Q4 [+,-] ",Q4_Q2_V,Q4_Q2_A)
```
