* タイムスタンプ処理

```python
# 関数定義
import json
from datetime import datetime

f = open('../timestamps.json', 'r')
stamps = json.load(f)['NGE9ynTJABg']
print(len(stamps))
print(stamps[0])

def calculate_dur(st, ed):
    _st = datetime.strptime(st, "%H:%M:%S")
    _ed = datetime.strptime(ed, "%H:%M:%S")
    dur = (_ed - _st).seconds
    return st, str(dur)
st1 , dur1 = calculate_dur(stamps[0][0],stamps[0][1])
print(dur1[:-1])
print(type(dur1))
```

```python
# メイン⭐⭐︎
# audio編集
import os
import ffmpeg
import json

Q1audio = os.listdir('EMOPIA_audio/')

with open('../timestamps.json', 'r') as f:
    stamps = json.load(f)
    notimestamp = []
    # audio読み込み
    for audio in Q1audio:
        # ファイル名読み込み
        if (str(audio[12:])=='mp3'):
            audio_name = str(audio[:-4])
            try:
                stamplist = stamps[audio_name]
                #　時間区切り＋拡張子変更   
                for i,time in enumerate(stamplist):
                    # タイムスタンプ区切り
                    print(time)
                    st1 , dur1 = calculate_dur(time[0],time[1])
                    print(dur1)
                    print(audio)
                    stream = ffmpeg.input('EMOPIA_audio/' +str(audio), ss=st1, t=dur1)
                    # 拡張子変更
                    stream = ffmpeg.output(stream, 'EMOPIA_audio_split/'+str(audio_name)+'_'+str(i)+'.mp3', format='mp3')
                    ffmpeg.run(stream)
            except:
                notimestamp.append(audio_name)

print(notimestamp)
f.close
```
